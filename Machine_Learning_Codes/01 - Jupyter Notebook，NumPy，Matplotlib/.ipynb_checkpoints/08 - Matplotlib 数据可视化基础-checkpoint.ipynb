### Matplotlib 基础

```python
# 一般的情况下主要是使用 matplotlib 中的子模块 pyplot 
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
# 绘制一个以 x 为横轴，以 y 为纵轴的图形
# 传入的第 1 个参数的值为横轴的值，传入的第 2 个参数的值为纵轴的值 
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
plt.plot(x, cosy)
```

```python
siny = y.copy()
```

```python
# 想要绘制两条曲线
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red')
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlim(-5, 15)
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
# plt.axis()的前两个数代表的是 x 轴的取值范围，后两个数代表的是 y 轴的取值范围
plt.axis([-1, 11, -2, 2]) 
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle = '--')
# plt.axis()的前两个数代表的是 x 轴的取值范围，后两个数代表的是 y 轴的取值范围
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的，曲线也是可以注明标签的
plt.plot(x, siny, label = 'sin(x)')
plt.plot(x, cosy, color = 'red', linestyle = '--', label = 'cos(x)')
# plt.axis()的前两个数代表的是 x 轴的取值范围，后两个数代表的是 y 轴的取值范围
plt.xlabel('x axis')
plt.ylabel('y value')
# plt.legend()表示在 show 的时候加上这个图示
plt.legend()
plt.show()
```

```python
# 想要绘制两条曲线，曲线的颜色是可以指定的,曲线的样式也是可以指定的
# 坐标轴的范围也是可以指定的，曲线也是可以注明标签的
plt.plot(x, siny, label = 'sin(x)')
plt.plot(x, cosy, color = 'red', linestyle = '--', label = 'cos(x)')
# plt.axis()的前两个数代表的是 x 轴的取值范围，后两个数代表的是 y 轴的取值范围
plt.xlabel('x axis')
plt.ylabel('y value')
# plt.legend()表示在 show 的时候加上这个图示
plt.legend()
# 给这个图加上一个标题
plt.title('Welcome to the ML World !')
plt.show()
```

### Scatter   Plot

```python
# 绘制散点图，第 1 个参数为横坐标的值，第 2 个参数为纵坐标的值
plt.scatter(x, siny)
plt.show()
```

```python
# 绘制散点图，第 1 个参数为横坐标的值，第 2 个参数为纵坐标的值，默认的颜色为蓝色
plt.scatter(x, siny)
plt.scatter(x, cosy, color = 'red')
plt.show()
```

```python
# 通常对于折线图来讲，plt.plot()横轴表示的是特征，纵轴表示的是特征的取值
# 对于散点图来讲，plt.scatter()横轴和纵轴都表示的是特征
# 一般绘制二维特征的时候使用散点图
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
```

```python
# 通常对于折线图来讲，plt.plot()横轴表示的是特征，纵轴表示的是特征的取值
# 对于散点图来讲，plt.scatter()横轴和纵轴都表示的是特征
# 一般绘制二维特征的时候使用散点图
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
```

```python
# 通常对于折线图来讲，plt.plot()横轴表示的是特征，纵轴表示的是特征的取值
# 对于散点图来讲，plt.scatter()横轴和纵轴都表示的是特征
# 一般绘制二维特征的时候使用散点图
# alpha表示不透明度，其中 0 表示完全透明的，1 表示完全不透明的，所以 0.5 表示的是半透明的
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha = 0.5)
```

```python
# 通常对于折线图来讲，plt.plot()横轴表示的是特征，纵轴表示的是特征的取值
# 对于散点图来讲，plt.scatter()横轴和纵轴都表示的是特征
# 一般绘制二维特征的时候使用散点图
# alpha表示不透明度，其中 0 表示完全透明的，1 表示完全不透明的，所以 0.5 表示的是半透明的
# 在中间的位置颜色非常的深，这是因为有许多个点叠加在一起的效果
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha = 0.1)
```
