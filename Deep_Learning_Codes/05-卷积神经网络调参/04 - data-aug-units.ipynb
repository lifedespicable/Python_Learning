```python
# resize
# crop
# flip
# brightness & contrast

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# resize 

# tf.image.resize_area
# tf.image.resize_bicubic
# tf.image.resize_nearest_neighbor

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 324, 432, 3])

resize_image = tf.image.resize_bicubic(img_decoded, [648, 864])

sess = tf.Session()
img_decoded_val = sess.run(resize_image)
img_decoded_val = img_decoded_val.reshape((648, 864, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box
# tf.image.crop_to_bounding_box
# tf.random_crop

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 324, 432, 3])

padded_image = tf.image.pad_to_bounding_box(
                        img_decoded, 50, 100, 450, 600)

sess = tf.Session()
img_decoded_val = sess.run(padded_image)
img_decoded_val = img_decoded_val.reshape((450, 600, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# flip
# tf.image.flip_up_down
# tf.image.flip_left_right
# tf.image.random_flip_up_down
# tf.image.random_flip_left_right

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 324, 432, 3])

flipped_image = tf.image.flip_up_down(img_decoded)

sess = tf.Session()
img_decoded_val = sess.run(flipped_image)
img_decoded_val = img_decoded_val.reshape((324, 432, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# brightness
# tf.image.adjust_brightness
# tf.image.random_brightness
# tf.image.adjust_contrast
# tf.image.random_contrast

name = './gugong.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1, 324, 432, 3])

new_image = tf.image.adjust_brightness(img_decoded, 0.5)

sess = tf.Session()
img_decoded_val = sess.run(new_image)
img_decoded_val = img_decoded_val.reshape((324, 432, 3))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```
